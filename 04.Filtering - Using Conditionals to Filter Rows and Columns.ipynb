```python
high_salary = (df['ConvertedComp'] > 70_000)

survey_countries = ['United States', 'United Kingdom', 'German', 'Canada']
```

```python
# Check if string contains the keyword 'Python' 
# 🧠 df[].str.contains('key') ⭐️

lang_python = df['LanguageWorkedWith'].str.contains('Python', na=False)  # ⭐️ Specify NaN as False!
```

```python
row_filter = (df['Country'].isin(survey_countries)) & high_salary & lang_python

col_filter = ['Country', 'LanguageWorkedWith', 'ConvertedComp']
```

```python
df.loc[row_filter, col_filter]
```
